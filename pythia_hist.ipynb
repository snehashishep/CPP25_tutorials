# Plots for $gg\to h$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Lepton multiplicity

### open the file, obtain the bin edges etc from the .dat file.

The first column of pythia output .dat file are bin centers, and the second column are events in the bins.

```python
filename = "../lep_mul-0.dat"  # Change this to your actual file name
data = np.loadtxt(filename)
```

```python
bin_centers = data[:, 0]
event_counts = data[:, 1]

# Convert bin centers to bin edges
# bin_width = np.diff(bin_centers).mean()  # Estimate bin width
bin_width = 0.5
# bin_edges = np.append(bin_centers - bin_width / 2, bin_centers[-1] + bin_width / 2)
bin_edges = np.append(bin_centers - 0.5/2, bin_centers[-1] + 0.5/2)
```

```python
# Plot histogram
plt.figure(figsize=(6,4))
plt.hist(bin_centers, bins=bin_edges, weights=event_counts, histtype='stepfilled', alpha=0.7, edgecolor="none")

# Labels and title
plt.xlabel(r"$n_{lep}$")
plt.ylabel("Events")
plt.title("lepton multiplicity")
plt.xlim(0,5)
plt.grid(True, alpha=0.3)

# Show plot
plt.show()
```

## $p_T$ of the leading two jets

```python
filename1 = "../jet_pt-0.dat"  # Change this to your actual file name
data1 = np.loadtxt(filename1)

filename2 = "../jet_pt-1.dat"  # Change this to your actual file name
data2 = np.loadtxt(filename2)

bin_centers = data1[:, 0]
event_counts1 = data1[:, 1]
event_counts2 = data2[:, 1]

# Convert bin centers to bin edges
# bin_width = np.diff(bin_centers).mean()  # Estimate bin width
bin_width = 10
bin_edges = np.append(bin_centers - bin_width / 2, bin_centers[-1] + bin_width / 2)
# bin_edges = np.append(bin_centers - 0.5/2, bin_centers[-1] + 0.5/2)
```

```python
# Plot histogram
plt.figure(figsize=(6,4))
plt.hist(bin_centers, bins=bin_edges, weights=event_counts1, histtype='step', alpha=0.7, edgecolor="blue", label='Jet 1')
plt.hist(bin_centers, bins=bin_edges, weights=event_counts2, histtype='step', alpha=0.7, edgecolor="red", label='Jet 2')

# Labels and title
plt.xlabel(r"$p_T$ [GeV]")
plt.ylabel("Events")
plt.title("jet $p_T$")
plt.xlim(10,500)
plt.grid(True, alpha=0.3)
plt.legend()

# Show plot
plt.show()
```
